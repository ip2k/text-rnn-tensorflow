```python
import requests
from bs4 import BeautifulSoup

bad_words = ['nword', 'nig', 'nio', 'noi', 'fag', 'cunt']  # lower-case, cases of chats will be converted for comparison
# portions of strings also work, as 'foo' will match both 'foo' and 'foobar' as well as 'foobarbaz' and 'quuxfoozxy'


def get_html(pg=1):
    resp = requests.get(f'http://stats.hyperiongaming.org/player_histories/chat/1186?chat_list_show_page={pg}&chat_list_showrows=500')
    if resp.status_code == 200:
        return resp.text
    else:
        raise Exception(f'Could not get player chat page {player_chat_page}')

        
def get_max_pg(html=''):
    soup = BeautifulSoup(html, 'html.parser')
    return int(soupg.find_all('span')[2].find_all('b')[1].text)  # third <span> and second <b> within that is the max page
   

def extract_chats(html=''):
    soup = BeautifulSoup(html, 'html.parser')
    filtered = soup.find_all('td', class_='t_sc')
    return [e.string for e in filtered]

 
        
if __name__ == '__main__':
    raw = get_html(pg=1)  # get the first page
    max_pg = get_max_pg(raw)  # figure out how many pages there are
    chats = extract_chats(raw)  # scrape
    if max_pg > 1:  # if there is >1 page of chats, get and scrape those, adding them to the chats list
        for curr_pg in range(2, max_pg):
            chats.extend(extract_chats(get_html(pg=curr_pg)))
    clean_chats = []
    dirty_chats = []
    for chat_raw in chats:
        chat = chat_raw.lower()  # lcase everything so we don't have to filter every permutation
        contains_badword = False
        for bad_word in bad_words:
            if bad_word in chat:
                contains_badword = True
                break
        if contains_badword:
            dirty_chats.append(chat_raw)
        else:
            clean_chats.append(chat_raw)
    long_chats = [c for c in clean_chats if len(c) > 20]
    print(long_chats)

    
```
